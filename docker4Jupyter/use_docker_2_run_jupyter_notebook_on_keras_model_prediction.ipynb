```python
import keras
from keras.models import load_model
m=load_model('pneumonia.h5')
#m.summary()
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
%matplotlib inline
im=cv2.imread('pneumonia.jpg')
im=cv2.resize(im,(150,150))
feed=im.reshape(1,150,150,3)
out=m.predict(feed)
title= "having Pneumonia " if np.argmax(out)==1 else "normal"
plt.imshow(im)
plt.title("prediction is : "+ title)
plt.show()
```
